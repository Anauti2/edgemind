```python
import numpy as np

def rotate(grid):
    return np.rot90(grid)

def flip(grid):
    return np.fliplr(grid)

def move_right(grid):
    new = np.zeros_like(grid)
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]-1):
            new[i][j+1] = grid[i][j]
    return new

PRIMITIVES = [rotate, flip, move_right]

def apply_program(grid, program):
    g = grid.copy()
    for f in program:
        g = f(g)
    return g

def simple_reasoning(grid):
    # demo reasoning: try simple transforms
    for func in PRIMITIVES:
        out = func(grid)
        if out.sum() != 0:
            return out, func.__name__
    return grid, "identity"
    
```
