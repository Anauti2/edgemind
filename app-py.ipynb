```python
import streamlit as st
import requests
import numpy as np

st.title("🧠 EdgeMind: Offline AI Reasoning")

text = st.text_input("Describe problem")

grid_input = st.text_area("Enter grid (e.g. [[1,0],[0,1]])")

if st.button("Solve"):
    try:
        grid = eval(grid_input)

        response = requests.post(
            "http://localhost:8000/solve",
            json={"grid": grid, "text": text}
        )

        data = response.json()

        st.write("### LLM Understanding")
        st.write(data["llm"])

        st.write("### Rule Used")
        st.write(data["rule_used"])

        st.write("### Output Grid")
        st.write(np.array(data["output"]))

    except:
        st.error("Invalid input")
        
```
